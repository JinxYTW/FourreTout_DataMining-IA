## Démonstration d'arbre de décision
Cette technique utilise les arbres de décision, qui permet d'utiliser de découper les données en sous-ensembles de données jusqu'à une limite.

### Importation des bibliothèques

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
import matplotlib.pyplot as plt
```

### Chargement du dataset Iris (Données d'iris récoltés en 1936)

```python
iris = datasets.load_iris()
```

### Préparation des données (récupération des données et des labels)

```python
X = iris.data  # Données (longueur et largeur des pétales, longueur et largeur des sépales )
y = iris.target # Labels (espèce de l'iris)
```

### Séparation des données 
On prend nos données, on découpe en deux parties :
- une partie entraînement : qui permet l'entrainement du modèle
- une partie test : qui permet de tester l'efficacité du modèle après entrainement

```python
# On sépare les données avec cette répartition : 70% pour l'entrainement et 30% pour le test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### Initialisation et entraînement du classfieur


```python
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
```

### Prédiction sur les données de test


```python
y_pred = clf.predict(X_test)
```

### Evaluation du modèle
On calcule la justesse du modèle

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"La justesse du classifieur en arbre de décision : {accuracy * 100:.2f}%")
```

### Visualisation des résultats
On va afficher l'arbre de décision afin de savoir les feuilles (qui sont les critères choisis par le classifier)

```python
plt.figure(figsize=(10, 8))
tree.plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names, rounded=True)
plt.show()
```

### Classification d'une nouvelle fleur
On va donner des valeurs afin de savoir à quelle classe sera attribuée.

```python
new_sample = [[5.5, 2.4, 3.8, 1.1]]  
predicted_class = clf.predict(new_sample)
print(f"The predicted class for the new sample is: {iris.target_names[predicted_class][0]}")
```

- `max_depth` : permettant de limiter la profondeur de l'arbre
- `min_samples_split` : le nombre minimum d'échantillions requis pour divisier un noeud interne.
